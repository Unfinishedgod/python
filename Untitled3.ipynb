####  1)정규방정식

![텍스트](https://render.githubusercontent.com/render/math?math=%5Chat%7B%5Cboldsymbol%7B%5Ctheta%7D%7D%20%3D%20%28%5Cmathbf%7BX%7D%5ET%20%5Cmathbf%7BX%7D%29%5E%7B-1%7D%20%5Cmathbf%7BX%7D%5ET%20%5Cmathbf%7By%7D&mode=inline)


```python
X_b = np.c_[np.ones((10, 1)), X]
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

![텍스트](https://render.githubusercontent.com/render/math?math=%5Chat%7By%7D%20%3D%20%5Cmathbf%7BX%7D%20%5Cboldsymbol%7B%5Chat%7B%5Ctheta%7D%7D&mode=inline)

```python
X_new = np.array([[160], [190]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.", markersize=20)
plt.show()
```

정규방정식을 사용하여 계산을 하면 최적의 선형 모델을 구할 수 있지만 피쳐가 늘어나면 계산 시간이 기하급수적으로 증가하는 한계점을 가지고 있다. sklearn의 LinearRegression을 사용하면 SVD 방법을 사용하여 계산하면 계산 복잡도는 O(n^2)으로 보다 줄어들지만 그래도 오래걸린다.

#### 경사 하강법의 스텝

![텍스트](https://render.githubusercontent.com/render/math?math=%5Cboldsymbol%7B%5Ctheta%7D%5E%7B%28%5Ctext%7Bnext%20step%7D%29%7D%20%3D%20%5Cboldsymbol%7B%5Ctheta%7D%20-%20%5Ceta%20%5Cdfrac%7B%5Cpartial%7D%7B%5Cpartial%20%5Cboldsymbol%7B%5Ctheta%7D%7D%20%5Ctext%7BMSE%7D%28%5Cboldsymbol%7B%5Ctheta%7D%29&mode=inline)

- 그림에서 J($ \theta $) 는 MSE 를 의미한다.
- MSE를 ML에서 Cost function, Object function이라고 부른다.
- $ \theta $ 값에 따라 비용함수의 값이 변하는데, 최종적으로 제일 아래부분에 도달하는 $ \theta $ 값을 찾아야 한다.
- 경사하강법에서는 $ \theta $ 와 b 에 대해 초기값을 지정하고 시작한다. 예를 들어 $ \theta $=5 와같다. 까만 점이 $ \theta $=5 라고 가정하면 그 부분에서의 기울기값 만큼 반대방향 으로 $ \theta $ 값을 바꾸어주면 된다.
- 이렇게 단계적으로 $ \theta $ 값을 바꿔가면서 비용값을 최소화 시키는 방법이 경사를 따라 내려가는 것과 비슷하다고 해서 경사하강법이란 이름으로 부른다.

```python
eta = 0.000003  # 학습률
n_iterations = 1000
m = 10

theta = np.random.rand(2,1)  # 랜덤 초기화

for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y.reshape(-1,1))
    theta = theta - eta * gradients
```

##### 경사 하강 시각화

```python
theta_path_bgd = []

def plot_gradient_descent(theta, eta, theta_path=None):
    m = len(X_b)
    plt.plot(X, y, "b.", markersize=20)
    n_iterations = 1000
    for iteration in range(n_iterations):
        if iteration < 10:
            y_predict = X_new_b.dot(theta)
            style = "b-" if iteration > 0 else "r--"
            plt.plot(X_new, y_predict, style)
        gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y.reshape(-1,1))
        theta = theta - eta * gradients
        if theta_path is not None:
            theta_path.append(theta)
    plt.xlabel("$x_1$", fontsize=18)
    plt.axis([160, 190, 60, 95])
    plt.title(r"$\eta = {}$".format(eta), fontsize=16)
```

```python
np.random.seed(42)
theta = np.random.rand(2,1)  # random initialization

plt.figure(figsize=(6,6))

plt.ylabel("$y$", rotation=0, fontsize=18)
plot_gradient_descent(theta, eta=0.00001, theta_path=theta_path_bgd)

plt.show()
```

#### Learning Rate

```python
np.random.seed(42)
theta = np.random.randn(2,1)  # random initialization

plt.figure(figsize=(20,6))
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.subplot(121); plot_gradient_descent(theta, eta=0.000028, theta_path=theta_path_bgd)
plt.subplot(122); plot_gradient_descent(theta, eta=0.000004)

plt.show()
```
